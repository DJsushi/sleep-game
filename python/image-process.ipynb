```python
from pathlib import Path
from PIL import Image
```

```python
# this code crops the images in the folder `imgs` and puts them into the `imgs-cropped` directory

# Define source and destination directories
source_dir = Path("imgs/")
dest_dir = Path("imgs-cropped/")

# Create destination directory if it doesn't exist
dest_dir.mkdir(exist_ok=True)

# Remove all existing images in destination directory
for file in dest_dir.glob("*"):
    if file.is_file():
        file.unlink()

# Process all image files
for image_path in source_dir.glob("*"):
    if image_path.is_file() and image_path.suffix.lower() in [
        ".jpg",
        ".jpeg",
        ".png",
        ".bmp",
        ".gif",
    ]:
        try:
            # Open the image
            img = Image.open(image_path)

            # Crop the image (adjust crop box as needed)
            # Format: (left, top, right, bottom)
            width, height = img.size
            crop_box = (0, 231, 1872, 1092)
            cropped_img = img.crop(crop_box)

            # Save cropped image
            output_path = dest_dir / image_path.name
            cropped_img.save(output_path)
            print(f"Cropped: {image_path.name}")
        except Exception as e:
            print(f"Error processing {image_path.name}: {e}")

print("Done!")
```
